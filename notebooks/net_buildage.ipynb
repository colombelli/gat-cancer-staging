```python
dt = df.T

col_1 = []
col_2 = []
col_r = []
col_p = []

for idx1, row1 in tqdm(dt.iterrows(), total=dt.shape[0]):
    for idx2, row2 in dt.loc[idx1:, :].iterrows():
        r, p = stats.pearsonr(row1.values, row2.values)
        col_1.append(idx1)
        col_2.append(idx2)
        col_r.append(r)
        col_p.append(p)
```

```python
corr_df = pd.DataFrame.from_dict({
        "sample1": col_1,
        "sample2": col_2,
        "r": col_r,
        "p": col_p
    })
```

```python
def filter_edge_list(df, r_filter, p_filter):
    return df.loc[(df.r >= r_filter) & (df['p'] <= p_filter)]
```

```python
filter_edge_list(corr_df, 0.7, 0.05)
```
