```python
import pandas as pd
from stellargraph import StellarGraph
import numpy as np

base_path = "/home/colombelli/Documents/datasets/acgt/kidney/"
patients_file = base_path + "igraph/patients_corr.csv"
gene_exp_file = base_path + "exp_log2.csv"
```

```python
df_patients = pd.read_csv(patients_file, sep=" ", header=None)
df_patients.columns = ["source", "target", "weight"]
df_patients
```

```python
sg = StellarGraph(edges=df_patients)
```

```python
print(sg.info())
```

```python
df_features = pd.read_csv(gene_exp_file, index_col=0)
df_features
```

```python
# Getting the index of the top n genes with the most variance among all genes
n = 10000
gene_variances = np.array(df_features.var())
top_genes = gene_variances.argsort()[-n:][::-1]

# Slicing feature dataframe with only the relevant genes
df_features = df_features.iloc[:, top_genes]
```

```python
# Getting only the patients present in the graph

df_features = df_features.loc[df_features.index.isin(list(df_patients['source'].values)+list(df_patients['target'].values))]
df_features
```

```python
# Testing to make sure that the slicing really considered the top_genes list as indexes for 
# the columns and not the column names itself

7369 in top_genes
```

```python
# Making StellarGraph object with the features and edges

sg = StellarGraph(df_features, df_patients)
print(sg.info())
```

```python
# Getting the cancer staging info

control = pd.read_csv(base_path+"normal.csv", usecols=[0]).iloc[:, 0].values
s1 = pd.read_csv(base_path+"split_stage/stage1.csv", usecols=[0]).iloc[:, 0].values
s2 = pd.read_csv(base_path+"split_stage/stage2.csv", usecols=[0]).iloc[:, 0].values
s3 = pd.read_csv(base_path+"split_stage/stage3.csv", usecols=[0]).iloc[:, 0].values
s4 = pd.read_csv(base_path+"split_stage/stage4.csv", usecols=[0]).iloc[:, 0].values
```

```python
# Reformating indexes changing tacks (-) for dots (.) 
control = [i.replace('-','.') for i in control]
s1 = [i.replace('-','.') for i in s1]
s2 = [i.replace('-','.') for i in s2]
s3 = [i.replace('-','.') for i in s3]
s4 = [i.replace('-','.') for i in s4]
```

```python
# Building class dataframe with staging information

class_info_dict = {
    "control": control,
    "stage1": s1,
    "stage2": s2,
    "stage3": s3,
    "stage4": s4
}

df_class = pd.DataFrame(columns=['id', 'class'])

for class_info, class_values in class_info_dict.items():
    df_temp = pd.DataFrame({'id': class_values, 'class': [class_info]*len(class_values)})
    df_class = df_class.append(df_temp, ignore_index=True)
       
df_class
```

```python
# Removing inexistent patients ids based on features dataframe
df_class.set_index('id', inplace=True)
df_class = df_class[df_class.index.isin(df_features.index)]
```

```python
df_features.index.name = 'id'
```

```python
# Saving the stellargraph processed data for later usage

df_patients.to_csv(base_path+"stellargraph/patients_edges.csv", index=False)
df_features.to_csv(base_path+"stellargraph/patients_features.csv")
df_class.to_csv(base_path+"stellargraph/patients_classes.csv")
```
